```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
celsius = np.array([-15.0, -87.0, -55.0, 38.0, 63.0, 12.0, -64.0, 77.0, -74.0, -46.0, 96.0, -95.0, -34.0, -97.0, 30.0, -79.0, -40.0, 54.0, 83.0, 89.0, 21.0, -76.0, -82.0, 45.0, -51.0],dtype = np.float32)
```

```python
fahrenheit = np.array([5.0, -124.6, -67.0, 100.4, 145.4, 53.6, -83.2, 170.6, -101.2, -50.8, 204.8, -139.0, -29.2, -142.6, 86.0, -110.2, -40.0, 129.2, 181.4, 192.2, 69.8, -104.8, -115.6, 113.0, -59.8],dtype = np.float32)
```

```python
X_train = celsius[:20]
X_test = celsius[20:]
y_train = fahrenheit[:20]
y_test = fahrenheit[20:]
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1,input_shape = [1])
])
model.compile(loss = tf.keras.losses.MAE,
              optimizer = tf.keras.optimizers.Adam(0.1),
              metrics = ['accuracy'])
history = model.fit(X_train,y_train,epochs = 500,verbose = 0)
```

```python
model.evaluate(X_test,y_test)
```

```python
pd.DataFrame(history.history).plot()
```
